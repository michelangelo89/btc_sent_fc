```python
excel_list = []

for f in files_xls:
    converted_df=pd.read_excel(f'/home/mrstoian/code/mrstoian/btc_sent_fc/raw_data/{f}', index_col=None,skiprows=6)
    converted_df.drop(converted_df.tail(4).index,inplace=True)
    converted_df.rename(columns={'Effective date ': 'date'}, inplace=True)
    converted_df['date'] = pd.to_datetime(converted_df['date'])
    converted_df.set_index('date',inplace=True)
    excel_list.append(converted_df)
    
financials_df = pd.concat(excel_list, axis=1)
```

```python
financials_df
```

## merge financials with manual joins

```python
final_financials_df = manual_joins.join(financials_df)
```
